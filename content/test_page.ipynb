# Adding Equations

How to add equations

There are fundamentally two ways to add equations.

__Method A__ Turn a cell into a Markdown cell and simply copy and paste your code as in LaTex.
For example, let's write _Maxwell's equations_

>```latex
\begin{aligned}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\
\nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0
\end{aligned}
```

If we copy and paste the above into a Markdown cell, and press Crtl+ENTER, then we get

\begin{align}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\
\nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0
\end{align}

# Itemising

Suppose we want to write a lists of elements. In LaTeX, we would usually do:

>```
\begin{itemize}
    \item My First item
    \item My Second item
    \item My Third item
\end{itemize}
```

Unfortunately, if we tried the two methods shown above, this will fail. Instead, we will get somehting ugly as shown below. 

\begin{itemize}
    \item My First item
    \item My Second item
    \item My Third item
\end{itemize}

````python
%%latex
\begin{itemize}
    \item My First item
    \item My Second item
    \item My Third item
\end{itemize}
````

__Markdown method__
In Markdown, we need to do instead

>```
* My First item
* My Second item
* My Third item
```

If you copy and paste the lines above in a Markdown cell, then you get

* My First item
* My Second item
* My Third item

Similarly, we can do something similar when enumerating items.

# Equation numbering/referencing

<span style="color:red">NOTE 1:</span> 
This is currently possible using the `MathJax Equation Numbering`. This can be installed according to instructions provided [here](https://stackoverflow.com/questions/41241984/equation-numbering-in-jupyter-notebooks/42940005#42940005).

A working example is shown next.

__First__ Say we want to number this equation

```latex
\begin{equation}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} = \frac{4\pi}{c}\vec{\mathbf{j}} 
\end{equation}
```

That produces this output

\begin{equation}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} = \frac{4\pi}{c}\vec{\mathbf{j}}
\label{eq:Maxwell_1}
\end{equation}

In _Markdown_ is not different. We can label equations like in _LaTeX_. So, we just add a label, like `\label{eq:Maxwell_1}`

```latex
\begin{equation}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} = \frac{4\pi}{c}\vec{\mathbf{j}}
\label{eq:Maxwell_1}
\end{equation}
```

__Second__ Now, suppose that you want to reference this equation in your text

In LaTeX, to reference the equation normally we do:

> Hey! Look at this pretty equation `\ref{eq:Maxwell_1}`

BUT, in Markdown we actually do:

> Hey! Look at this pretty equation `$\eqref{eq:Maxwell_1}$`

which procudes a hyperlink to the referenced equation.

> Hey! Look at this pretty equation \eqref{eq:Maxwell_1}

## **Method X** Using HTML attributes

__FIRST__ Convert `.bib` file into a formated `.html` file using [`nbconvert`](https://github.com/jupyter/nbconvert-examples). 

This approach allows us to insert citations in Markdown cells using HTML data attributes.

Installation instructions for `nbconverter` can be found [here](https://nbconvert.readthedocs.io/en/latest/install.html). This can be easily installed with `conda` though, a number of dependencies need to be considered. See instructions.

_Example_: We want to turn a file like [this](https://github.com/broncio123/riemann_book/blob/master/riemann.bib) into something like [this](https://github.com/broncio123/riemann_book/blob/master/riemann_bib.html)

For instance, a `bib` entry 

```latex
@article{greenberg1997analysis,
  title={Analysis and approximation of conservation laws with source terms},
  author={Greenberg, JM and Leroux, AY and Baraille, R and Noussair, A},
  journal={SIAM Journal on Numerical Analysis},
  volume={34},
  number={5},
  pages={1980--2007},
  year={1997},
  publisher={SIAM}
}

```

is turned into an `html` object like this

```html
<a name="greenberg1997analysis"></a><pre>

  author = {Greenberg, JM and Leroux, AY and Baraille, R and Noussair, A},
  journal = {SIAM Journal on Numerical Analysis},
  volume = {34},
  number = {5},
  pages = {1980--2007},

  publisher = {SIAM}
}
</pre>
```

within an HTML file.
